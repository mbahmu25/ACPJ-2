```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
gambar = cv2.imread("foto.png")

plt.imshow(gambar)
```

```python
cv2.imshow("Tes",gambar)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hsv = cv2.cvtColor(gambar,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

```python
plt.imshow(gambar[:,:,0],cmap="Reds_r")
plt.title("Red")
```

```python
plt.imshow(gambar[:,:,1],cmap="Greens_r")
plt.title("Green")
```

```python
plt.imshow(gambar[:,:,2],cmap="Blues_r")
plt.title("Biru")
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
  histr = cv2.calcHist([gambar],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0,256])
plt.title("Histogram")
plt.show()
```

```python
plt.imshow(hsv[:,:,0],cmap="gray")
plt.title("Hue")
```

```python
plt.imshow(hsv[:,:,1],cmap="gray")
plt.title("Saturation")
```

```python
plt.imshow(hsv[:,:,2],cmap="gray")
plt.title("Value")
```

```python
color = ('r','g','b')
for i,col in enumerate(color):
  histr = cv2.calcHist([hsv],[i],None,[360],[0,360])
  
  plt.plot(histr,color = col)
plt.legend(["Hue","Value","Saturation"])
#   plt.xlim([0,256])
# plt.title("Histogram")
# plt.show()
```

```python
bawah = np.array([100,100,100])
atas = np.array([255,255,255])
mask = cv2.inRange(gambar[:,180:490,:],bawah,atas)
res = cv2.bitwise_and(gambar[:,180:490,:],gambar[:,180:490,:],mask=mask)
plt.imshow(res)
```

```python
plt.subplot(1,3,1)
plt.imshow(res[:,:,0],cmap="Reds_r")
plt.subplot(1,3,2)
plt.imshow(res[:,:,1],cmap="Greens_r")
plt.subplot(1,3,3)
plt.imshow(res[:,:,2],cmap="Blues_r")
```
